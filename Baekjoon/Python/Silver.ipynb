#### 문제

---

루트 없는 트리가 주어진다. 이때, 트리의 루트를 1이라고 정했을 때, 각 노드의 부모를 구하는 프로그램을 작성하시오.
<br><br>
#### 입력

---

첫째 줄에 노드의 개수 N (2 ≤ N ≤ 100,000)이 주어진다. 둘째 줄부터 N-1개의 줄에 트리 상에서 연결된 두 정점이 주어진다.
<br><br>
#### 출력

---

첫째 줄부터 N-1개의 줄에 각 노드의 부모 노드 번호를 2번 노드부터 순서대로 출력한다.

````python
n = int(input())

linked = [[] for i in range(n)]
for i in range(n - 1) :
  n1, n2 = map(int,input().split(" "))

  linked[n1 - 1].append(n2)
  linked[n2 - 1].append(n1)

tree = {}
tree[1] = None

while (len(tree.keys()) < n) :
  for i in range(n) :
    for j in linked[i] :
      if i + 1 in tree.keys() and j not in tree.keys():
        tree[j] = i + 1
      elif i + 1 not in tree.keys() and j in tree.keys():
        tree[i + 1] = j

for i in range(2, n + 1) :
  print(tree[i])
````
